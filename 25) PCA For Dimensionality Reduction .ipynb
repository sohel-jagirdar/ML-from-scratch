```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('glass.data')
data
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
X=data.drop(['index','Class'],axis=1)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(X)
```

```python
X=pd.DataFrame(data=scaled_data)
X
```

```python
X.describe()
```

```python
from sklearn.decomposition import PCA
model=PCA()
Principal_component=model.fit_transform(X)
plt.figure()
# Plotting Explain Variance Ratio
plt.plot(np.cumsum(model.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

```python
pca=PCA(n_components=4)
pca_data=pca.fit_transform(X)
new_pca_DF=pd.DataFrame(data=pca_data,columns = ['principal component 1', 'principal component 2','principal component 3','principal component 4'],)
```

```python
new_pca_DF
```

# Visualizing the Principal components Using Random Sample Data

```python
# Creating sample data for the Principal components Visualization
np.random.seed(1)
X = np.dot(np.random.random(size=(2, 2)), np.random.normal(size=(2, 200))).T
#Visualizing the 2D Data
plt.plot(X[:, 0], X[:, 1], 'o')
plt.axis('equal');
```

```python
#Fitting PCA with 2 Principal Component For Check Variance
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_)
print(pca.components_)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

```python
#Again Fitting PCA with 95% Of Variance
clf = PCA(0.95) # keep 95% of variance
X_trans = clf.fit_transform(X)
print(X.shape)
print(X_trans.shape)
```

```python
#Dark Blue Line Represent PCA1 having 95% of Variance Ratio
X_new = clf.inverse_transform(X_trans)
plt.plot(X[:, 0], X[:, 1], 'o', alpha=0.2)
plt.plot(X_new[:, 0], X_new[:, 1], 'ob', alpha=0.8)
plt.axis('equal');
```

### Coverting High Dimensional Data To Low Dimensional Data For Visualization By Using PCA

```python
from sklearn.datasets import load_digits
digits = load_digits()
X = digits.data
y = digits.target

pca = PCA(2)  # project from 64 to 2 dimensions
Xproj = pca.fit_transform(X)
print(X.shape)
print(Xproj.shape)
```

```python
# Creating a scatter plot of the datapoints
plt.scatter(Xproj[:, 0], Xproj[:, 1], c=y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar();
```

## Creating PCA For Cancer Detection Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
df=load_breast_cancer()
data=pd.DataFrame(data=df.data,columns=df['feature_names'])
```

```python
data
```

```python
df.target
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
S_scaler=StandardScaler()
X=S_scaler.fit_transform(data)
X1=X.copy()
```

```python
X1=pd.DataFrame(X)
X1.describe()
```

```python
from sklearn.decomposition import PCA
pca_model=PCA()
pca_data=pca_model.fit_transform(X1)
pca_data
```

```python
plt.plot(np.cumsum(pca_model.explained_variance_ratio_))
plt.xlabel('No. Of PCA')
plt.ylabel('Explained Variance Ratio')
plt.title('(fig. E.V.R )')
```

## As We See in Fig.E.V.R it clearly show that  Only 5 PCA conatain 90% Of Variance

```python
model_2d=PCA(n_components=2)
X2=model_2d.fit_transform(X)
X2.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X2[:,0],X2[:,1],c=df['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
model_2d.components_
```

##### In this numpy matrix array, each row represents a principal component, and each column relates back to the original features. we can visualize this relationship with a heatmap:

```python
df_comp = pd.DataFrame(model_2d.components_,columns=df['feature_names'])
df_comp
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```

#### This heatmap and the color bar basically represent the correlation between the various feature and the principal component itself.
