```python
X = data.drop(['income'], axis=1)
y = data['income']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
train_x.shape, test_x.shape
```

```python
category=[var for var in categorical if var !='income']
category
```

```python
train_x[category].isnull().mean()
```

```python
for var in category:
    if train_x[var].isnull().mean()>0:
        print(var,train_x[var].isnull().mean())
```

```python
# impute missing categorical variables with most frequent value
for df in [train_x , test_x]:
    df['workclass'].fillna(train_x['workclass'].mode()[0] , inplace=True)
    df['occupation'].fillna(train_x['occupation'].mode()[0], inplace=True)
    df['native_country'].fillna(train_x['native_country'].mode()[0], inplace=True)
    
```

```python
import category_encoders as ce
encoder=ce.OneHotEncoder(cols=category)
train_x=encoder.fit_transform(train_x)
test_x=encoder.transform(test_x)
```

```python
cols=train_x.columns
```

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
train_x=scaler.fit_transform(train_x)
test_x=scaler.transform(test_x)
```

```python
train_x=pd.DataFrame(train_x,columns=cols)
test_x=pd.DataFrame(test_x,columns=cols)
```

```python
train_x
```

```python
## Random Forest Classifier model with default parameters 
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=0)
rfc.fit(train_x,train_y)
y_pred=rfc.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy With Default RFC : ' ,accuracy_score(test_y,y_pred))
```

```python
rfc_100=RandomForestClassifier(n_estimators=100,random_state=0)
rfc_100.fit(train_x,train_y)
y_pred_rfc_100=rfc_100.predict(test_x)
```

```python
print('Accuracy With n_estimators 100 RFC : ' ,accuracy_score(test_y,y_pred_rfc_100))
```

### finding Important Feature 

```python
clf=RandomForestClassifier(random_state=0)
clf.fit(train_x,train_y)
```

```python
feature_scores = pd.Series(clf.feature_importances_, index=train_x.columns).sort_values()
```

```python
f, ax = plt.subplots(figsize=(30, 24))
ax = sns.barplot(x=feature_scores, y=feature_scores.index, data=data)
ax.set_title("Visualize feature scores of the features")
ax.set_yticklabels(feature_scores.index)
ax.set_xlabel("Feature importance score")
ax.set_ylabel("Features")
plt.show()
```

```python
train_x = train_x.drop(['native_country_41'], axis=1)

test_x = test_x.drop(['native_country_41'], axis=1)
```

```python
rfc_final=RandomForestClassifier(n_estimators=100,random_state=0)
rfc_final.fit(train_x,train_y)
y_pred_final=rfc_final.predict(test_x)
```

```python
print('Accuracy After Performing Import Feature : ',accuracy_score(test_y,y_pred_final))
```
