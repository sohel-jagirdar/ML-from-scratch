```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from sklearn.datasets import load_boston
from sklearn.ensemble import GradientBoostingRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
data=load_boston()
```

```python
df=pd.DataFrame(data=data.data,columns=data.feature_names)
```

```python
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    print(df[col].isnull().sum())
```

```python
df['taget']=data.target
```

```python
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
train_x
```

```python
from sklearn.metrics import r2_score
gbr=GradientBoostingRegressor(random_state=0)
gbr=gbr.fit(train_x,train_y)
y_pred_test=gbr.predict(test_x)
y_pred_train=gbr.predict(train_x)
```

```python
print('Train Score With Defoult Params:',r2_score(train_y,y_pred_train))
print('Test Score With Defoult Params:',r2_score(test_y,y_pred_test))
```

### We Can See Our data Is Ovrfitted To the Train Model
### Now We Can Perform Hyperparameter Tuning for Finding Best Params


```python
from sklearn.model_selection import GridSearchCV
params={
    'loss':['ls'],
    'learning_rate':[0.01,0.1],
    'n_estimators':[100,150,200,250],
    'criterion':['friedman_mse','mse',],
    'max_depth':[3,4]
}
gscv_gbr=GridSearchCV(estimator=GradientBoostingRegressor(),param_grid=params,cv=3)
gscv_gbr.fit(train_x,train_y)
```

```python
print(gscv_gbr.best_params_,gscv_gbr.best_score_)
```

```python
GBR=GradientBoostingRegressor(n_estimators = 150,criterion= 'friedman_mse',learning_rate= 0.1, loss= 'ls', max_depth= 3 )
GBR.fit(train_x,train_y)
y_pred_test=GBR.predict(test_x)
y_pred_train=GBR.predict(train_x)
```

```python
print('Train Score With Defoult Params:',r2_score(train_y,y_pred_train))
print('Test Score With Defoult Params:',r2_score(test_y,y_pred_test))
```
