```python
#importing Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import statsmodels.api as sm
```

```python
# Importing Dataset
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Multiple-Linear-Regression/master/50_Startups.csv')
df.head()
```

```python
#Spliting Dataset into X,y
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X
```

```python
#Appying OHE (One Hot Encoding)
cities=pd.get_dummies(df.State,drop_first=True)
cities.head()
```

```python
X=pd.concat([X,cities],axis=1)
X
```

```python
X=X.drop(["State"],axis=1)
X
```

```python
X.head()
```

### Checking Correlation and Statistic Analysis Using StaticModel.API

```python
X1=sm.add_constant(X)
X1.head()
```

```python
model=sm.OLS(y,X1).fit()
model.summary()
```

```python
#checking corr() with other data
X1.iloc[:,1:].corr()
```

```python
X.head()
```

```python
#splittin data into train & test 
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=.2,random_state=0)
```

```python
#fitting Data Into Model
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
lin.fit(X,y)
```

```python
#predicting train_x and test_x
y_train_pred=lin.predict(train_x)
y_test_pred=lin.predict(test_x)
```

```python
from sklearn.metrics import r2_score
#Accuracy For Training 
print(r2_score(train_y,y_train_pred))
#Accuracy For Testing
print(r2_score(test_y,y_test_pred))
```
