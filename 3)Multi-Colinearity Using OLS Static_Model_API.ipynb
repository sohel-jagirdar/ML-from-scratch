## Example 1: 

```python
# importing static model
import statsmodels.api as sm
```

```python
import pandas as pd
import matplotlib.pyplot
```

```python
# loading database from github
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Multicollinearity/master/data/Advertising.csv')
X=df.iloc[:,1:-1]
y=df.iloc[:,-1]
y
```

```python
X
```

### Adding Constant to X 

```python
X=sm.add_constant(X)
X
```

```python
# OLS (ordnary Least Square Model fitting using y,X
model=sm.OLS(y,X).fit()
```

```python
model.summary()
```

```python
X.iloc[:,1:].corr()
```

## Example 2 :

```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Multicollinearity/master/data/Salary_Data.csv')
df.head()
```

```python
X=df.iloc[:,:2]
y=df.iloc[:,-1]
```

```python
# Adding Coeficiant to X
X=sm.add_constant(X)
X
```

```python
model=sm.OLS(y,X).fit()
```

```python
model.summary()
```
